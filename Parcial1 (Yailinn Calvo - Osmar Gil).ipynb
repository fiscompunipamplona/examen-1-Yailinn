$(n,k) = \frac{n!}{k!(n-k)!}$

```python
#Inciso 1
from math import* 

print('En los valores ingresados n>k')
n=int(input("Ingrese el valor de n: "))
k=int(input("Ingrese el valor de k: "))
def binomial(n,k):
    if k==0:
        ecuacion=1
    elif k>=1:
        ecuacion=(factorial(n))/(factorial(k)*factorial(n-k))
    return(ecuacion)

a=binomial(n,k)
print('EL valor del coeficiente binomial es: ',int(a))
```
